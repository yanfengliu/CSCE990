```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
import os
```

```python
root_dir ='C:/Users/yliu60/Documents/GitHub/CSCE990/output'
out = cv2.VideoWriter('test.avi',cv2.VideoWriter_fourcc(*'DIVX'), 30, (416, 128*2))
count = 0
filelist = os.listdir(os.path.join(root_dir, 'depth'))
filelist.sort(key=lambda f: int(f.split('.')[0]))
for filename in filelist:
    if count < 2000:
        board = np.zeros((128*2, 416, 3))
        image = cv2.imread(os.path.join(root_dir, 'color', filename))
        depth = cv2.imread(os.path.join(root_dir, 'depth', filename))
        board[:128, :, :] = image
        board[128:, :, :] = depth
        board = board.astype(np.uint8)
        out.write(board)
        count += 1
    else:
        break
out.release()
```

```python
for filename in filelist:
    if count < 2000:
        color = cv2.imread(os.path.join(root_dir, 'color', filename))
        # color = cv2.cvtColor(color, cv2.COLOR_RGB2BGR)
        depth = cv2.imread(os.path.join(root_dir, 'depth', filename))
        depth = cv2.cvtColor(depth, cv2.COLOR_RGB2BGR)
        board = np.zeros((128, 416*2, 3))
        board[:, :416, :] = color
        board[:, 416:, :] = depth
        board = board * 255
        board = board.astype(np.uint8)
        # need to be BGR format
        out.write(board)
        count += 1
        print(count)
    else:
        break
out.release()
```
